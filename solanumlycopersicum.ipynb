```python
df2['Name'] = np.nan
df2['Note'] = np.nan
df2['Other'] = np.nan
df2
```

```python
for i in range(df2.shape[0]):
    if type(df2['Annotation'][i]) is str:
        fields = df2['Annotation'][i].split(';')
        other = []
        for f in fields:
            if f.startswith('Name'):
                name_s = f[5:]
                df2['Name'][i] = name_s
            elif f.startswith('Note'):
                note_s = f[5:]
                df2['Note'][i] = note_s
            else:
                other.append(f)

        if len(other) > 0:
            other_s = ';'.join(other)
            df2['Other'][i] = other_s
```
